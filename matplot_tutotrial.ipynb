```python
# plotting multiple subplot in one using subplot method
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import NullFormatter
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1=np.arange(1,6,1)
t2=np.arange(1,4,3)

# figure 1
plt.figure(1)
plt.title('My first figure')

# subplot 1 in figure 1
plt.subplot(211)
plt.plot(t1,f(t1),'bs')

# subplot 2 in figure 1
plt.subplot(212)
plt.plot(t2,f(t2),'r^')

#figure 2
plt.figure(2)
plt.title('My second figure')

# subplot 1 in figure 2
plt.subplot(311)
plt.plot(t1,t1,'bs')
plt.subplot(312)

# subplot 2 in figure 2
plt.plot(t2,f(t2),'r^')

plt.gca().yaxis.set_minor_formatter(NullFormatter())
plt.subplots_adjust(top=1,bottom=0.008,left=0.10,right=0.95,
                    hspace=0.25,wspace=1)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import NullFormatter

np.random.seed(19680801)

y=np.random.normal(loc=0.5,scale=0.4,size=1000)
y=y[(y>0)&(y<1)]
y.sort()
x=np.arange(len(y))

#figure1
plt.figure(1)

#linear use plt.yscale('linear')
plt.subplot(221)
plt.plot(x,y)
plt.xscale('linear')
plt.yscale('linear')
plt.title('linear')
plt.grid(True)  # grid is for lining inside graph line visualization

# log
plt.subplot(222)
plt.plot(x,y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

#symmetric log use .yscale('symlog')
plt.subplot(223)
plt.plot(x,y-y.mean())  #using mean function
plt.yscale('symlog')
plt.title('symlog')
plt.grid(True)

#logit
plt.subplot(224)
plt.plot(x,y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.gca().yaxis.set_minor_formatter(NullFormatter())
# Adjusting the subplot layout
#as they merge into one another cuz of no space in between them
plt.subplots_adjust(top=0.92,bottom=0.08,left=0.10,right=0.95,
                    hspace=0.25,wspace=0.35)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# build a rectangle in axes coords
left,width=.25,.5
bottom,height=.25,.5
right=left+width
top=bottom+height

# create figure
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
# axes coordinates are 0,0 is bottom left and 1,1
# is upper right

p=patches.Rectangle((left,bottom),width,height,fill=False,
  transform=ax.transAxes,clip_on=False)

ax.add_patch(p)

ax.text(left,bottom,'left top',horizontalalignment='left',
       verticalalignment='top',
       transform=ax.transAxes)

ax.text(left,bottom,'left bottom',
       horizontalalignment='left',verticalalignment='bottom',
       transform=ax.transAxes)

ax.text(right,bottom,'center top',
       horizontalalignment='center',verticalalignment='top',
       transform=ax.transAxes)

ax.text(left,0.5*(bottom+top),'right center',
       horizontalalignment='right',verticalalignment='center',
       rotation='vertical',transform=ax.transAxes)

ax.text(left,top,'rotated\n with newlines',
       horizontalalignment='center',
       verticalalignment='center',
       rotation=45,
       transform=ax.transAxes)

ax.text(right,0.5*(bottom+top),'centered',
       horizontalalignment='center',
       verticalalignment='center',
       rotation='vertical',
       transform=ax.transAxes
       )

ax.text(0.5*(left+right),0.5*(bottom+top),'middle',
        horizontalalignment='center',verticalalignment='center',
        fontsize=20,color='red',transform=ax.transAxes)

ax.set_axis_off()
plt.show()
```

```python
# blended transformations 
# to make bleneded lines and spans we can use in-built functions 
# axhline(),axvline(),axxhspan(),axvspan()
# can also implement horizontal span using blended tranform 
# blended transform can only be used for seperable transformations like cimple cartesian 
# coordinates but not on inseperable transformations like Poloar Transforms
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches 
import matplotlib.transforms as transforms
fig=plt.figure()
ax=fig.add_subplot(111)
x=np.random.randn(1000)
ax.hist(x,30)
ax.set_title(r'$\sigma=1 \/ \dots\/ \sigma=2$',fontsize=16)
# x coords of this transformation are data and  y coords are axes
trans=transforms.blended_transform_factory(ax.transData,ax.transAxes)
#highlight 1..2 stddev region with span
# we want x to be data coordinates and y to span from 0..1 in axes coords
rect=patches.Rectangle((1,0),width=1,height=1,transform=trans,color='yellow',alpha=0.5)
ax.add_patch(rect)
plt.show()
# blended transformation where x is in data coords and y in axes coords is useful
# and we 've mplbuses it internally for drawing ticks and ticklabels
# The methods arematplotlib.axes.Axes.get_xaxis_transform()
# and matplotlib.axes.Axes.get_yaxis_transform().
# blended_tranform_factory() can be replaced by get_xaxis_transform()
# syntax::   trans=ax.get_xaxis_transform()
```
